```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset manually skipping the first 4 rows using iloc
data = pd.read_csv(r"D:\40 project data sets\task 1 prodigy\API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv", skiprows=4)
# df_cleaned = df_raw.iloc[4:].reset_index(drop=True)  # second way to skip the first 4 rows
data.head()
```

```python
# Step 3: Select data for a specific year (e.g., 2020)
latest_year = '2020'  # Modify if necessary
data_2020 = data[['Country Name', latest_year]].dropna()
data_2020.columns = ['Country', 'Population']  # Rename columns
```

```python
# Step 4: Sort by population in descending order
data_2020 = data_2020.sort_values(by='Population', ascending=False)
```

```python

# Step 5: Create a Bar Chart for the top 10 most populated countries
plt.figure(figsize=(12, 6))
sns.barplot(x='Population', y='Country', data=data_2020.head(10), palette='coolwarm')
plt.xlabel('Population', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title(f'Top 10 Most Populated Countries in {latest_year}', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Step 6: Create a Histogram for world population distribution
plt.figure(figsize=(10, 5))
# Specifies the number of classes in the distribution (the larger the number, the more accurate the analysis).
sns.histplot(data_2020['Population'], bins=30, kde=True, color='blue')
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title(f'World Population Distribution in {latest_year}', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
